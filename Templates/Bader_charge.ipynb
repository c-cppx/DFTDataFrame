```python
CO_atoms = Adsorbateatoms.loc[CO.index]
O_atoms = Adsorbateatoms.loc[O.index]
H_atoms = Adsorbateatoms.loc[H.index]
OH_atoms = Adsorbateatoms.loc[OH.index]
```

```python
from pandas import Series
 
def Bader_ref(Frame, dic):
    baderref = Series(index=Frame.index)
    for key, value in dic.items():
        baderref.loc[Frame[Frame['Symbols'] == key].index] = value
    return baderref
```

```python
CO_atoms['bader_ref'] = Bader_ref(CO_atoms, {'Ni': 10 , 'Ga': 3, 'Cu': 11, 'Zn':12, 'C':2.20, 'O':7.7962}) # from gas phase CO
O_atoms['bader_ref'] = Bader_ref(O_atoms, {'Ni': 10 , 'Ga': 3, 'Cu': 11, 'Zn':12, 'O':6}) # from gas phase O
OH_atoms['bader_ref'] = Bader_ref(OH_atoms, {'Ni': 10 , 'Ga': 3, 'Cu': 11, 'Zn':12, 'C':0, 'O':7, 'H':0}) # from gas phase OH
H_atoms['bader_ref'] = Bader_ref(H_atoms, {'Ni': 10 , 'Ga': 3, 'Cu': 11, 'Zn':12, 'H':1}) # from gas phase H2
```

```python
CO_atoms['Bader_change'] = -(CO_atoms['Bader_charge'] - CO_atoms['bader_ref'])       
O_atoms['Bader_change'] = -(O_atoms['Bader_charge'] - O_atoms['bader_ref'])       
OH_atoms['Bader_change'] = -(OH_atoms['Bader_charge'] - OH_atoms['bader_ref'])      
H_atoms['Bader_change'] = -(H_atoms['Bader_charge'] - H_atoms['bader_ref'])    
```

```python
Adsorbateatoms2 = pd.concat([CO_atoms, OH_atoms])
Adsorbateatoms2 = pd.concat([Adsorbateatoms2, H_atoms])
Adsorbateatoms2 = pd.concat([Adsorbateatoms2, O_atoms])
Adsorbateatoms2
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import numpy as np

def D3plot(x,y,z,values,title, elev=0, azim=120, s=400, lw=1.25, edgecolor='black', font=10, labels=True):
    # setup the normalization and the colormap
    normalize = mcolors.Normalize(vmin=-6, vmax=6)
    colormap = cm.seismic
    fig = plt.figure(figsize=(15,8))
    ax = fig.add_subplot(projection='3d',title=title)
    ax.scatter(x,y,z, c=colormap(list(map(normalize, values))), s = s, alpha=1, linewidth=lw, edgecolor=edgecolor)
    if labels:
        for i in np.arange(0, len(values)):
            i = int(i)
            if np.abs(values[i])>0.2:
                ax.text(x[i],y[i], z[i], round(values[i],1), ha="center" ,va='center', fontsize=font)
    ax.view_init(elev=elev, azim=azim)
    # setup the colorbar
    scalarmappaple = cm.ScalarMappable(norm=normalize, cmap=colormap)
    scalarmappaple.set_array(values)
    plt.colorbar(scalarmappaple, shrink=0.5, ax=ax)
    ax.axis('off')
    ax.set_aspect('equal')
    plt.tight_layout()#rect=[0, 0,1, 1])
    #plt.savefig('+title+'.svg', bbox_inches='tight')
    plt.show()
```
